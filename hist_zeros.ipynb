Le but de ce code est de tracer un histogramme des zéros pour le rapport de mi-parcours (simple).


```python
# ...existing code...
import pandas as pd
import matplotlib.pyplot as plt

path = "/Users/romain/Desktop/FINAL_GRAVITY_TRAINING_MATRIX (1).csv"
df = pd.read_csv(path)
n_rows = len(df)
zeros = 199000 - n_rows

print("n_rows =", n_rows)
print("zeros =", zeros)

plt.figure(figsize=(6,4))
plt.bar(['zeros'], [zeros], color='C0')
plt.ylabel('Count')
plt.title('Nombre de zéros (199000 - n_rows)')
plt.tight_layout()
plt.show()
# ...existing code...
```
